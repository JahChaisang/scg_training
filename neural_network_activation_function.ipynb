## Sigmoid Activation Function

```python
import numpy as np
import math

def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
sigmoid(0.45)
```

```python
x = np.linspace(-5, 5, 100)
```

```python
y = sigmoid(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid")
plt.plot(x, y)
plt.plot([-6,6], [0.5, 0.5], 'r')
```

### Relu Activation Function

```python
def identity(x):
    return x.copy()

def relu(x):
    y = x.copy()
    y[x<0] = 0
    return y
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

x = np.linspace(-2.5, 2.5, 100)
y = relu(x)
sns.set_style("darkgrid")
plt.plot(x, y)
plt.gca().set_aspect('equal', adjustable='box')
```

```python
y = identity(x)
sns.set_style("darkgrid")
plt.plot(x, y)
plt.gca().set_aspect('equal', adjustable='box')
```

```python
y = np.tanh(x)
sns.set_style("darkgrid")
plt.plot(x, y)
plt.gca().set_aspect('equal', adjustable='box')
```
