### Load data and split them to X and Y

```python
import numpy as np
data = np.loadtxt('data.txt', delimiter=',')
m = data.shape[0]
```

```python
data.shape
```

```python
o = np.ones((m, 1))
x = np.hstack((o, data[:, 0:2]))
y = data[:, -1]
```

```python
n = x.shape[1]
```

### Gradient descent algorithm

```python
def grad(x, y, theta):
    gradient = np.zeros(n)
    pred = np.dot(x, theta)
    error = pred - y
    for i in range(n):
        gradient[i] = np.sum(x[:, i] * error)/m
    return gradient

def costfunc(x, y, theta):
    pred = np.dot(x, theta)
    error = pred - y
    return np.sum(error ** 2) / (2 * m)

alpha = 0.001
iters = 5000

theta = np.random.rand(n)
for i in range(iters):
    gradient = grad(x, y, theta)
    theta = theta - alpha * gradient
    if i % 100 == 0:
        print("Gradient Descent cost = " + str(costfunc(x, y, theta)))

print("Gradient Descent: the parameters")
print(theta)
print("Gradient Descent cost = " + str(costfunc(x, y, theta)))
```
